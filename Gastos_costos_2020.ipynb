```python
import pandas as pd
import numpy as np
```

```python
#Eliminar primera fila o filas innecesarias
gastos_costos_20=pd.read_excel("Gastos y costos 20-23.xlsx",sheet_name="2020", skiprows=5)
gastos_costos_20
```

```python
data1_gastos20=gastos_costos_20.copy()
```

```python
gastos_costos_20["FOLIO"]=gastos_costos_20["FOLIO"].fillna("F0000")
gastos_costos_20
```

```python
gastos_costos_20["GASTO"]=gastos_costos_20["GASTO"].fillna("INDEFINIDO")
gastos_costos_20
```

```python
gastos_costos_20["TC"]=gastos_costos_20["TC"].fillna("ffill")
gastos_costos_20
```

```python
gastos_costos_20["IMPORTE"]=gastos_costos_20["IMPORTE"].fillna(round(gastos_costos_20['IMPORTE'].mean(),1))
gastos_costos_20
```

```python
gastos_costos_20["IVA"]=gastos_costos_20["IVA"].fillna(round(gastos_costos_20['IVA'].mean(),1))
gastos_costos_20
```

```python
gastos_costos_20["POLIZA"]=gastos_costos_20["POLIZA"].fillna("SIN_POLIZA")
gastos_costos_20
```

```python
valores_nulos_gastos20=gastos_costos_20.isnull().sum()
valores_nulos_gastos20
```
