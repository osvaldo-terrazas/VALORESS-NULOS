```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ventas_totales=pd.read_csv("Ventas_totales.csv")
ventas_totales.head(7)
```

```python
ventas_totales_null=ventas_totales.isnull().sum()
ventas_totales_null
```

```python
#Realizamos una copia del dataframe
ventas_totales_copy = ventas_totales.copy()
```

```python
#Sustitucion por promedio para valores numericos
ventas_totales_copy["salon_ventas"]=ventas_totales_copy["salon_ventas"].fillna(round(ventas_totales["salon_ventas"].mean(),1))
ventas_totales_copy["tarjetas_debito"]=ventas_totales_copy["tarjetas_debito"].fillna(round(ventas_totales["tarjetas_debito"].mean(),1))
ventas_totales_copy["tarjetas_credito"]=ventas_totales_copy["tarjetas_credito"].fillna(round(ventas_totales["tarjetas_credito"].mean(),1))
ventas_totales_copy["otros_medios"]=ventas_totales_copy["otros_medios"].fillna(round(ventas_totales["otros_medios"].mean(),1))
ventas_totales_copy["subtotal_ventas_alimentos_bebidas"]=ventas_totales_copy["subtotal_ventas_alimentos_bebidas"].fillna(round(ventas_totales["subtotal_ventas_alimentos_bebidas"].mean(),1))
ventas_totales_copy["bebidas"]=ventas_totales_copy["bebidas"].fillna(round(ventas_totales["bebidas"].mean(),1))
ventas_totales_copy["almacen"]=ventas_totales_copy["almacen"].fillna(round(ventas_totales["almacen"].mean(),1))
ventas_totales_copy["panaderia"]=ventas_totales_copy["panaderia"].fillna(round(ventas_totales["panaderia"].mean(),1))
ventas_totales_copy["lacteos"]=ventas_totales_copy["lacteos"].fillna(round(ventas_totales["lacteos"].mean(),1))
ventas_totales_copy["carnes"]=ventas_totales_copy["carnes"].fillna(round(ventas_totales["carnes"].mean(),1))
ventas_totales_copy["verduleria_fruteria"]=ventas_totales_copy["verduleria_fruteria"].fillna(round(ventas_totales["verduleria_fruteria"].mean(),1))
ventas_totales_copy["alimentos_preparados_rotiseria"]=ventas_totales_copy["alimentos_preparados_rotiseria"].fillna(round(ventas_totales["alimentos_preparados_rotiseria"].mean(),1))
ventas_totales_copy["indumentaria_calzado_textiles_hogar"]=ventas_totales_copy["indumentaria_calzado_textiles_hogar"].fillna(round(ventas_totales["indumentaria_calzado_textiles_hogar"].mean(),1))
ventas_totales_copy["electronicos_articulos_hogar"]=ventas_totales_copy["electronicos_articulos_hogar"].fillna(round(ventas_totales["electronicos_articulos_hogar"].mean(),1))
ventas_totales_copy["otros"]=ventas_totales_copy["otros"].fillna(round(ventas_totales["otros"].mean(),1))
```

```python
ventas_totales_null_1=ventas_totales_copy.isnull().sum()
ventas_totales_null_1
```

```python
#Convertir a archivos csv
ventas_totales_copy.to_csv("VENTAS TOTALES CORREGIDO.csv")
```
